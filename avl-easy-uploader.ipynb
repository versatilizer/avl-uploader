```python
from Google import Create_Service
from googleapiclient.http import MediaFileUpload
from apiclient.errors import HttpError

import requests
from requests.exceptions import HTTPError

import numpy as np # calculation
import pandas as pd # Excel for Python
import os
import time
from datetime import datetime

#importing database to work with
avldb = 'avl-clean-db.csv'
data = pd.read_csv(avldb, sep=',')

#normalizing data
data['video_name'] = data['video_name'].astype(str)
data['download_time'] = pd.to_datetime(data['download_time'], format = '%d/%m/%Y %H:%M:%S')
data['upload_time'] = pd.to_datetime(data['upload_time'], format = '%d/%m/%Y %H:%M:%S')

#iterating through empty rows
idx = 0
for idx in range(len(data)):
    #downloading the video file from the AVLoops website and saving it
    url = data.iloc[idx]['video-url']
    video_name = url.rsplit('/', 1)[1]
    if data.iloc[idx]['upload_time'] is pd.NaT:
        r = requests.get(url, allow_redirects=False)
        open(video_name, 'wb').write(r.content)
        data.at[idx,'video_name'] = video_name
        now = datetime.now().strftime('%d/%m/%Y %H:%M:%S')
        data.at[idx,'download_time'] = now

        #downloading the video file from the AVLoops website and saving it
        url = data.iloc[idx]['video-url']
        r = requests.get(url, allow_redirects=False)
        video_name = url.rsplit('/', 1)[1]
        open(video_name, 'wb').write(r.content)
        
        #logging file name and download time
        data.at[idx,'video_name'] = video_name
        now = datetime.now().strftime('%d/%m/%Y %H:%M:%S')
        data.at[idx,'download_time'] = now

        #creating video description
        description = 'Download original animation: '+data.iloc[idx]['productpage-link-href']+'. ''AVLoops is a visual content marketplace for HD, 2D and 3D animations specifically designed for electronic dance music. ''You can search through thousands of loops that are available in our growing library contributed by talented content producers from all around the world.'

        #upload video
        CLIENT_SECRET_FILE = 'client_secrets.json'
        API_NAME = 'youtube'
        API_VERSION = 'v3'
        SCOPES = ['https://www.googleapis.com/auth/youtube.upload']

        service = Create_Service(CLIENT_SECRET_FILE, API_NAME, API_VERSION, SCOPES)

        request_body = {
            'snippet': {
                'categoryI': 1,
                'title': data.iloc[idx]['item-title'],
                'description': description,
                'tags': ['Animation', 'AV Loops', 'Abstract', 'DJ Loops']
            },
            'status': {
                'privacyStatus': 'private',
                #'publishAt': upload_date_time,
                'selfDeclaredMadeForKids': False, 
            },
            'notifySubscribers': False
        }

        mediaFile = MediaFileUpload(data.iloc[idx]['video_name'])
        try:
            response_upload = service.videos().insert(
                part='snippet,status',
                body=request_body,
                media_body=mediaFile
            ).execute()

            #log succesful upload time
            now2 = datetime.now().strftime('%d/%m/%Y %H:%M:%S')
            data.at[idx,'upload_time'] = now2

            #delete the downloaded file on successful upload
            os.remove(data.iloc[idx]['video_name'])
      
            
data.to_csv(f'avl-clean-db-2021-01-02-test.csv', index = False
            #, mode='a', header=False
           )
```
